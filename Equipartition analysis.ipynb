## A closer look at equipartition (minimum energy)

The dependencies of total energy in electrons $E_e$ and in magnetic field $E_B$ on source magnetic field form the conceptual basis of the minimum energy argument. In chapter 1 we provided a quick and easy formula for this, in the following we explore the theoretical origins and derive a more accurate formula.

We will assume that we observe (or can infer, assume) a radio spectrum with some spectral index $\alpha$ between upper and lower frequencies $\nu_1, \nu_2$. We will need the integrated radio luminosity between these limits, and not just the specific radio luminosity at a certain frequency.

$$L=4 \pi D^2 \int_{\nu_1}^{\nu_2} F_{\nu} d\nu = 4 \pi D^2 F_{\nu_2} \nu_2^{-\alpha} \left( \frac{\nu_2^{\alpha+1} - \nu_1^{\alpha+1}}{\alpha+1} \right) $$

(an equivalent formula can of course be constructed using $\nu_1$ as the reference frequency). For a flat spectrum ($\alpha = 0$) source the expression is very simple

$$L = 4 \pi D^2 F_{\nu} (\nu_2 - \nu_1)$$ which is why the radio luminosity of flat spectrum sources is often simply estimated as $L = 4 \pi D^2 \nu F_{\nu}$ (for example in most works on the radio:X-ray correlation in X-ray binaries, and the fundamental plane of black hole activity).

**Remember that in our choice of definitions, $\alpha = (1-p)/2$.**

Note also, in comparing the analysis here with the simplified (but relatively accurate) formlua of Longair et al., that the $L_{\nu}$ used in that formula is the 'specific' luminosity in units erg s$^{-1}$ Hz$^{1}$ whereas the *integrated* luminosity used here and defined above is in erg s$^{-1}$ (and is therefore a larger quantity).

### Energy in magnetic fields

This one is very straightforward: the energy density in magnetic fields is $B^2 / 8\pi$ (remember we are working entirely and consistently in cgs units, this is **not** the expression in SI units), therefore for a volume $V$ the total energy in B fields is

$$E_B = \frac{B^2}{8 \pi} f V = \frac{B^2}{8 \pi} f \frac43 \pi R^3 = \frac{ fB^2 R^3}6$$

where $f$ is a *filling factor* which represents the fraction of the physical source's volume which is actually occupied by the magnetic field and relativistic particles.

### Energy in electrons

The total number in electrons, for a power law distribution of index $-p$ between energy limits $E_1$ and $E_2$ is given by:

$$N = \int^{E_2}_{E_1} N_0 E^{-p} dE = \left[ \frac{N_0 E^{1-p}}{1-p} \right]^{E_2}_{E_1} = \frac{N_0}{1-p} (E_2^{1-p}-E_1^{1-p}) $$ 

The total energy in these electrons is given by:

$$E_e = \int^{E_2}_{E_1} E \times N_0 E^{-p} dE = \left[ \frac{N_0 E^{2-p}}{2-p} \right]^{E_2}_{E_1} = \frac{N_0}{2-p} (E_2^{2-p}-E_1^{2-p}) $$

The rate of energy loss in the form of synchrotron radiation from an electron of energy $E$ in a magnetic field $B$ is given by:

$$-\frac{dE}{dt} = c_2 B^2 E^2$$

Therefore the *luminosity* observed from this spectrum of particles in a field $B$ is given by:

$$L=-\int^{E_2}_{E_1} \frac{dE}{dt} N(E) dE =  N_0 c_2 B^2 \int^{E_1}_{E_2} E^{2-p} dE = \frac{N_0 c_2 B^2}{3-p}(E_2^{3-p}-E_1^{3-p})  $$

Performing the above integrations, eliminating for $N_0$, and making $E_e$ the subject, we get:

$$E_e = c_2^{-1} L B^{-2} \frac{(p-3)}{(p-2)} \frac{(E_1^{2-p}-E_2^{2-p})}{(E_1^{3-p}-E_2^{3-p})}$$

For a given electron energy and magnetic field, we can characterise the synchrotron emission as coming from a single frequency $\nu = c_1 B E^2$, i.e. $E = \nu^{1/2} c_1^{-1/2} B^{-1/2}$. So, we can write:

$$E_e = c_2^{-1} c_1^{1/2} \tilde{c}(p, \nu_1, \nu_2) B^{-3/2} L = c_{12}(p,\nu_1,\nu_2) B^{-3/2} L$$

where

$$c_{12} = c_2^{-1} c_1^{1/2} \tilde{c}(p, \nu_1, \nu_2)$$

and

$$\tilde{c}(p, \nu_1, \nu_2) = \frac{(p-3)}{(p-2)} \frac{\nu_1^{(2-p)/2}-\nu_2^{(2-p)/2}}{\nu_1^{(3-p)/2}-\nu_2^{(3-p)/2}}$$

The pseudo-constant $c_{12}$ encapsulates all the information about the observed (or interpolated, extrapolated, assumed) frequency range of the observations and the slope of the energy spectrum between those limits. 

**The key point to take onboard here is that the total energy in electrons has a $B^{-3/2}$ dependence**. Since $E_B \propto B^2$, the total energy is given by

$$E_{e+B} = E_e + E_B =  const. B^{-3/2} + const. B^2$$

and therefore *there must be a minimum of the total energy as a function of $B$* (assuming there are no other contributions to the energy budget which have a different dependence on $B$).

This minimum occurs at *equipartition* of the energy, when $E_e$ and $E_B$ are almost equal (differentiate $E_{e+B}$ with respect to $B$):

$$E_B = \frac34 \eta E_e$$

where $\eta$ indicates how much energy is stored in the protons which accompany the electrons (in the case of a normal baryonic plasma); $\eta = 1$ is generally considered for the minimum energy condition.

Substituting in our definitions of $E_B$ and $E_e$ at equipartition we get

$$\frac{B^2 f R^3}{6} = \frac34 \eta c_{12} B^{-3/2} L$$

which leads to

$$B_{eq} = (4.5)^{2/7} (\eta c_{12} L)^{2/7} f^{-2/7} R^{-6/7}$$

which is the equipartition magnetic field.

At equipartition, the total minimum energy is

$$E_{min} = \frac 74 \eta E_e = \frac74 \eta c_{12} B^{-3/2} L$$

and substituting in $B = B_{eq}$ we get

$$E_{min} =  0.92 \eta^{4/7} c_{12}^{4/7} f^{3/7} R^{9/7} L^{4/7}$$

(in Pacholczyk the constants and $c_{12}^{4/7}$ are amalgamated into a new constant, $c_{13}$, but it is not really necessary)

Finally, we have the full, and readily calculable, expression for the minimum energy and associated equipartition magnetic field which we can associate with any synchrotron emitting region for which we observe a luminosity $L$, in a given frequency range (encapsulated in $c_{12}$) and can estimate a size $R$.


```python
import matplotlib
import seaborn
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
import scipy.constants

c1=6.27e18
c2=2.37e-3
# speed of light 
c=3.0e10
# relativistic sound speed of c/sqrt[3]
cs=c/(3.0**0.5)

def sizefromt(t,vexp):
    r=t*vexp
    V=(4.0/3.0)*np.pi*(r**3.0)
    return r,V

def LumTot(snu2,D,nu1,nu2,alpha): 
# here s is the flux density at the upper frequency nu2
    L2 = 4 * np.pi * ((D * 3.1e21)**2.0) * (snu2 * 1e-26) * (nu2**(-alpha)) * ((nu2**(1+alpha))-(nu1**(1+alpha))) / (1+alpha)
    print ('Calculated total radiative luminosity'),('L ='),('%.2e' % L2),(' erg / sec ')
    return L2

#### USER INPUTS HERE ####
p=2.5
alpha=(1.0-p)/2.0
(r,V)=sizefromt(300.0,c)

# flux density of peak in mJy
s=50.0
# distance in kpc
d=11.0
# observing frequency range
nu1=5.0e7
nu2=14.0e9
# eta is measure of total energy in particles compared to energy in electrons
eta=1.0
# f is filling factor
f=0.1
####

print ('Input: p='),(p),(' s='),(s),('mJy, D='),(d),('kpc, nu(1,2)=('),(nu1/1e9) ,(','),(nu2/1e9),(') GHz')
print ('Size from rise time r='),('%.2e' % r),(' cm, Filling factor = '),f
L=LumTot(s,d,nu1,nu2,alpha)

ctildetop=(nu1**((2.0-p)/2.0))-(nu2**((2.0-p)/2.0))
ctildebottom=(nu1**((3.0-p)/2.0))-(nu2**((3.0-p)/2.0))
ctilde=((p-3.0)/(p-2.0))*(ctildetop/ctildebottom)
c12=(c2**-1.0)*(c1**0.5)*ctilde

# as an extra check you might want to print out c12 for 
# comparison with Pacholcyzk table 8 to confirm this implemented correctly
# print ('Pacholcyzk variable c12 ='),('%.2e' % c12)

beq=((4.5 * eta * c12 * L)**(2.0/7.0))*(f**(-2.0/7.0))*(r**(-6.0/7.0))

print ('Equipartition field B_eq='),('%.2e' % beq),(' G')

# we can calculate minimum energy using either of the formulae we conclude with above
# feel free to check they give the same answer

emin1=(7.0/4.0)*eta*c12*L*(beq**(-1.5))
emin2=0.92*((eta*c12*L)**(4.0/7.0))*(f**(3.0/7.0))*(r**(9.0/7.0))

eb=(beq**2.0)*(r**3.0)/6.0
ee=c12 * (beq**(-1.5)) * L

#print 'Corresponding minimum energy1=',('%.2e' % emin1),' erg'
print ('Corresponding minimum energy ='),('%.2e' % emin2),(' erg')

print ('( Energy in electrons ='),('%.2e' % ee),('erg, Energy in magnetic field ='),('%.2e' % eb),(')')
```

```python
# We can finish this by plotting how the Energy 
# will vary around this minimum with varying B
# this is the same code as in chapter 1

E=emin2
x=np.linspace(beq/100,beq*100,10000)

Ee=E*(x/beq)**(-1.5)
Eb=E*(x/beq)**(2.0)
Etot=Ee+Eb

plt.figure(figsize=(12,8))
plt.yscale('log')
plt.xscale('log')
plt.plot(x,Ee,'r',label="Electrons")
plt.plot(x,Eb,'b',label="Magnetic Field")
plt.plot(x,Etot,'g', label="Total Energy")
plt.xlabel('B (Gauss)')
plt.ylabel('Energy (erg)')
plt.title('Energy as a function of B')
plt.legend()
# comment / uncomment the next line in order to output image to a file
plt.savefig("Min_E.png")
###
plt.show()
plt.close()
```
